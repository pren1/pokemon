```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
import time as t
```

```python
image = cv2.imread('001.jpg')
plt.imshow(image)
plt.show()
cv_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(cv_rgb)
plt.show()
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray)
plt.show()
canny = cv2.Canny(gray, 30, 150)
canny = np.uint8(np.absolute(canny))
canny = cv2.bitwise_not(canny)
plt.imshow(canny,cmap='gray')
plt.show()
cv2.imwrite('canny.jpg', canny)
```

```python
image = cv2.imread('001.jpg')

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
sobelx = cv2.Sobel(gray, cv2.CV_64F, 1, 0,ksize = 3)
sobely = cv2.Sobel(gray, cv2.CV_64F, 0, 1,ksize = 3)

sobelx = np.uint8(np.absolute(sobelx))
sobely = np.uint8(np.absolute(sobely))
sobelcombine = cv2.bitwise_or(sobelx,sobely)
sobelcombine = cv2.bitwise_not(sobelcombine)
plt.imshow(sobelcombine,cmap='gray')
plt.show()
cv2.imwrite('sobel.jpg', sobelcombine)
```

```python
image = cv2.imread('001.jpg')

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
lap = cv2.Laplacian(gray, cv2.CV_64F,ksize=1)
lap = np.uint8(np.absolute(lap))

lap = cv2.bitwise_not(lap)
plt.imshow(lap,cmap='gray')
plt.show()
cv2.imwrite('lap.jpg', lap)
```
